```python
from scipy import misc
import matplotlib.pyplot as plt
#charger la photo via scipy
face = misc.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
```

```python
h = face.shape[0]
w = face.shape[1]

#zoomer d'un quart sur l'image
zoom_face = face[h//4 : -h//4, w//4 : -w//4]
plt.imshow(zoom_face, cmap=plt.cm.gray)
plt.show()
```

```python
#rendre les couleurs sombre en blanc
zoom_face[zoom_face > 150] = 255
plt.imshow(zoom_face, cmap=plt.cm.gray)
plt.show()
```

```python
#compresser l'image
zoom_face = zoom_face[::2, ::2]
plt.imshow(zoom_face, cmap=plt.cm.gray)
plt.show()
```

```python
#compresser l'image initiale
face = face[::2, ::2]
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
```
