```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.describe()
```

```python
titanic.dropna(axis=0, inplace=True)
```

```python
titanic.describe()
```

```python
titanic.head()
```

```python
sns.pairplot(titanic)
# pour mieux visualiser : utiliser catplot et boxplot pour explorer le categorical plots
## on a les allignement et tt, parceque on a beaucoup de variable catgorique soit 0 ou 1
##les 3 classes des passagers, ....
## pour ca on va utiliser categorical plot
```

```python
sns.catplot(x='pclass', y='age', data=titanic, hue='sex')
# voir la repartition des H et F selon la classe et l'age
```

```python
sns.boxplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
#distribution
sns.distplot(titanic['fare'])

#conclu : la majorit ont voyag avec un billet moin cher
```

```python
sns.jointplot('age', 'fare', data=titanic, kind='kde') #kernel density
```

```python
sns.heatmap(titanic.corr()) #mat de correlation
```

```python
sns.heatmap(titanic.corr())
```
