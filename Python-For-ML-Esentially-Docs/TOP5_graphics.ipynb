```python
#top 5 graphiques matplotlib

# 5 - Scatter classification
### classification avec plt scatter
### par exemple : pour classeur un mail en tant que spam ou pas

    # practice : Iris flower
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

x = iris.data
y = iris.target
names = list(iris.target_names)

print(f' x contient {x.shape[0]} exemple et {x.shape[1]} variables')
print(f' il y a {np.unique(y).size} classes')
```

```python
#pour visualiser : utilisation de plt.scatter
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5, s=x[:,2]*100) 
# c : couleur / c=y en fonction des classes de y (3)
# alpha : transparence des points
# s : tailles des points / s=100 / s=[x:2]*100 : en fonction du 1er et 2eme attributs (largeur+longueur petal)
plt.xlabel('longueur sépal')
plt.ylabel('largeur sépal')
```

```python
# graphique 3D : sub package matplotlib : toolkits

from mpl_toolkits.mplot3d import Axes3D
ax = plt.axes(projection='3d')  #creation objet ax avec lequel travailler : POO
ax.scatter(x[:, 0], x[:, 1], x[:,2], c=y, alpha=0.5, s=x[:,2]*30)
```

```python
#visualiser des fonction mathematiques

f = lambda x, y: np.sin(x) + np.cos(x+y)*np.cos(x)
X = np.linspace(0, 5, 100)
Y = np.linspace(0 ,5, 100)
X, Y = np.meshgrid(X, Y)

Z = f(X, Y)

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')
```

```python
# 3 : histogram : data distributed
#reprend les donénes de IRIS

plt.hist(x[:, 0], bins=20) # sur x : longueur du sépal / y : nombre d'apparition du sépal
# créer plusieur histogrammes pour toutes les attributs
plt.hist(x[:, 1], bins=20)
```

```python
#histogram 2D
plt.hist2d(x[:, 0], x[:, 1]) #cmap='blues'
plt.xlabel('longueur sépal')
plt.ylabel('largeur sépal')

plt.colorbar()
```

```python
# 2 : CONTOUR PLOT

plt.contour(X, Y, Z, 20, colors='black') #20: beaucoup de contour
```

```python
#astuce
plt.contourf(X, Y, Z, 20, cmap='RdGy') #cmap : pour colorer
plt.colorbar()
```

```python
#practice : on IRIS


iris = load_iris()
x = iris.data
y = iris.target
x
```

```python
n = x.shape[1] # nombre de colonnes
plt.figure(figsize=(12,8))

for i in range(n):
  plt.subplot(n//2, n//2, i+1)
  plt.scatter(x[:, 0], x[:, i], c=y)
  plt.xlabel('0')
  plt.ylabel(i)
  plt.colorbar(ticks=list(np.unique(y)))

plt.show()
```
