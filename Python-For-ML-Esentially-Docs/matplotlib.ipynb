```python
#methode pyplot

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12, 8))#dimensions, 
x = np.linspace(0, 50, 2000)
plt.subplot(2, 1, 1)
plt.plot(x,np.cos(x))
plt.plot(x, x)
plt.subplot(2, 1, 2)
plt.plot(x, x**2, c='red', label='quadratique')
plt.plot(x, np.sin(x), label='sinus x')
plt.show()
plt.savefig('figure.png')
```

```python
# methode OOP

fig , ax = plt.subplots(2, 1, sharex=True)
w = np.linspace(0, 50, 2000)
ax[0].plot(w, w**5)
ax[1].plot(w, np.sin(w))
plt.show()
```
